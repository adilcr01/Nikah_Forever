```python
df_temp=df.copy()
```

```python
df_temp['permanent_address']=df_temp['permanent_address'].apply(lambda x: x.strip("[").strip(']').strip("{").strip("}").strip("'").strip("'"))
```

```python
df_temp.head()
```

```python
permanent_country_new=[]
permanent_city_new=[]
permanent_state_new=[]
permanent_postal_code_new=[]
permanent_address_new=[]
for i in df_temp['permanent_address']:
    for j in range(0,5):
        if j==0:
            try:
                permanent_country_new.append(i.split(',')[j].split(':')[1].split('"')[1])
            except:
                permanent_country_new.append('null')
        elif j==1:
            try:
                permanent_city_new.append(i.split(',')[j].split(':')[1].split('"')[1])
            except:
                permanent_city_new.append('null')
        elif j==2:
            try:
                permanent_state_new.append(i.split(',')[j].split(':')[1].split('"')[1])
            except:
                permanent_state_new.append('null')
        elif j==3:
            try:
                permanent_postal_code_new.append(i.split(',')[j].split(':')[1].split('"')[1])
            except:
                permanent_postal_code_new.append('null')
        
        elif j==4:
            try:
                permanent_address_new.append(i.split(',')[j].split(':')[1].split('"')[1])
            except:
                permanent_address_new.append('null')
            
        else:
            print('error')
```

```python
df_new_col=pd.DataFrame(data={'permanent_country':permanent_country_new,'permanent_state':permanent_state_new,'permanent_city':permanent_city_new})
```

```python
df_new_col.head()
```

```python
plt.pie(df_lucknow['membership'].value_counts(),labels=df_lucknow['membership'].unique(),autopct='%.2f')
plt.show()
```
